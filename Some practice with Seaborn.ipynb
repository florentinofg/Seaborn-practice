## Seaborn

Seaborn is a graphic library built on top of Matplotlib. It allows to make your charts prettier, and facilitates some of the common data visualisation needs.

<a id="library"></a>
<h1 style='background:SteelBlue; border:0; color:white'><center>Libraries</center></h1>

```python
# Import dependencies

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

<h1 style='background:SteelBlue; border:0; color:white'><center>Load Data</center></h1>

```python
olympic = pd.read_csv('athlete_events.csv')
healthcare=pd.read_csv('train_data.csv')
covid_india=pd.read_csv('covid_19_india.csv')
google_playstore=pd.read_csv('googleplaystore.csv')
iris=pd.read_csv('Iris.csv')
corona_virus = pd.read_csv('covid_19_data.csv')
youtube=pd.read_csv('USvideos.csv')
shootings=pd.read_csv('shootings.csv')
```

<a id="color_scatter_plot"></a>
<h2 style='border:3px solid SteelBlue; color:SteelBlue'><center>Using color Palletes in Scatter Plot</center></h2>

```python
olympic.head()
```

```python
# using sequencial color scheme
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='Blues')
plt.title('Age vs. Height')
```

```python
# using diverging color scheme
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='PuOr')
plt.title('Age vs. Height')
```

```python
# using discrete color scheme
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='Set1')
plt.title('Age vs. Height')
```

<a id="seaborn_themes"></a>
<h1 style='background:SteelBlue; border:0; color:white'><center>Seaborn Themes</center></h1>
<a href="#TOC" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover">Go to TOC</a>

Seaborn has five built-in themes to style its plots: `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`. Seaborn defaults to using the darkgrid theme for its plots, but you can change this styling to better suit your presentation needs.

To use any of the preset themes pass the name of it to `sns.set_style()`.

<a id="darkgrid"></a>
<h2 style='border:3px solid SteelBlue; color:SteelBlue'><center>Seaborn Dark Grid Theme</center></h2>

```python
sns.set_style("darkgrid")
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title("Seaborn Dark Grid Theme")
```

<a id="dark"></a>
<h2 style='border:3px solid SteelBlue; color:SteelBlue'><center>Seaborn Dark Theme</center></h2>

```python
sns.set_style("dark")
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title("Seaborn Dark Theme")
```

<a id="whitegrid"></a>
<h1 style='border:3px solid SteelBlue; color:SteelBlue'><center>Seaborn White Grid Theme</center><h1>

```python
sns.set_style("whitegrid")
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title("Seaborn White Grid Theme")
```

<a id="white"></a>
<h1 style='border:3px solid SteelBlue; color:SteelBlue'><center>Seaborn White Theme</center><h1>
